```python
import numpy as np
import math
from matplotlib import pyplot as plt
```

```python
#tridiagonal solver from hw 4
def tridiag(A,b):
    n = len(A)
    U = np.zeros((n,n))
    L = np.zeros((n,n))
    y = np.zeros((n,1))
    x = np.zeros((n,1))
    U[0][0] = A[0][0]
    for j in range (0,n-1):
        L[j+1][j] = A[j+1][j]/U[j][j]
        U[j+1][j+1] = (A[j+1][j+1])-(L[j+1][j]*A[j][j+1])
        U[j][j+1] = A[j][j+1]
        L[j][j] = 1
    L[n-1][n-1]=1
    y[0][0] = b[0][0]
    for i in range (1,n):
        y[i][0] = b[i][0] - L[i][i-1]*y[i-1][0]     
    x[n-1][0]=y[n-1][0]/U[n-1][n-1]
    for k in range (n-2,-1,-1):
        x[k][0] = (y[k][0]-A[k][k+1]*x[k+1][0])/U[k][k]
    return(x)     
```

```python
def jacobi(N):
    #create matricies A and b based on given boundary value probelm
    #code adapted from hw 4
    A = np.zeros((N+1,N+1))
    b = np.zeros((N+1,1))
    h = 1/N
    A[0][0]=1 
    A[N][N]=1
    for i in range(1,N):
        A[i][i-1] = -1/(h**2)
        A[i][i] = (2/(h**2))+(math.pi**2)
        A[i][i+1] = -1/(h**2)
        b[i][0] = 2*(math.pi**2)*math.sin(math.pi*(i/N))
        
    #create M
    M = np.zeros((N+1,N+1))
    x = np.zeros((N+1,1))
    tol = h*0.1
    iterations = 0 
    for j in range (0,N+1):
        M[j][j] = A[j][j]
        
    #create N matrix
    NN = -1*(A-M)
    
    #plug into equation to solve for x
    for i in range (0,N**N):
        x_i = x
        c = np.dot(NN,x_i)+b
        x_ii = tridiag(M,c)
        g = np.linalg.norm(b-np.dot(A,x_i))
#check tolerance
        if np.linalg.norm(b-np.dot(A,x_ii))<= tol:
            x = x_ii
            break

        else:
            x = x_ii
            iterations += 1
    return(x, iterations)  
```

```python
def GS(N):
    #create matricies A and b based on given boundary value probelm
    #code adapted from hw 4
    A = np.zeros((N+1,N+1))
    b = np.zeros((N+1,1))
    h = 1/N
    A[0][0]=1 
    A[N][N]=1
    for i in range(1,N):
        A[i][i-1] = -1/(h**2)
        A[i][i] = (2/(h**2))+(math.pi**2)
        A[i][i+1] = -1/(h**2)
        b[i][0] = 2*(math.pi**2)*math.sin(math.pi*(i/N))
        
    #create M
    M = np.zeros((N+1,N+1))
    x = np.zeros((N+1,1))
    tol = h*0.1
    iterations = 0 
    for j in range (0,N+1):
        for i in range (0,j+1):
            M[j][i] = A[j][i]
            
    #create N matrix
    NN = -1*(A-M)

    #plug into equation to solve for x
    for i in range (0,N**N):
        x_i = x
        c = np.dot(NN,x_i)+b
        x_ii = tridiag(M,c)
        if np.linalg.norm(b-np.dot(A,x_ii))<= tol:
            x = x_ii
            break
        else:
            x = x_ii
            iterations += 1

    return(x, iterations)
        
```
